```python
import numpy as np

from gwg import ising

from torch.distributions import Distribution
Distribution.set_default_validate_args(False)
```

```python
seeds = np.arange(8)
sigmas = np.arange(0.25, 0.51, 0.05)
temps = ['bg-1', 'bg-2', 'bg-5', 'hb-10-1', 'gwg', 'gwg-3', 'gwg-5']
```

```python
for seed in seeds:
    for sigma in sigmas:
        args = {
            'save_dir': f'save/ising/10x10/{seed}/{np.round(sigma, 2)}', # Directory to save results
            'n_steps': 100000, # Number of steps for the sampler
            'n_test_samples': 100, # Number of samples
            'seed': seed, # Random seed
            'dim': 10, # Size of the Ising lattice
            'sigma': sigma, # Connectivity strength
            'bias': 0, # External field
            'save_sample_every': 10, # Save frequency
            'print_every': 10000, # Print frequency
            'subsample': 1, # Subsampling factor
            'burn_in': 0.1, # Burn-in period for ESS
            'ess_statistic': 'hamming', # Statistic for ESS
            'no_ess': False, # Flag to disable ESS
        }
        ising.ising_sample(args, temps)
```
