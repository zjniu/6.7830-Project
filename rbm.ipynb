```python
import numpy as np

from gwg import rbm

from torch.distributions import Distribution
Distribution.set_default_validate_args(False)
```

```python
seeds = np.arange(8)
temps = ['bg-1', 'bg-2', 'hb-10-1', 'gwg', 'gwg-3', 'gwg-5']
```

```python
for seed in seeds:
    args = {
        'model_path': f'save/rbm/mnist/{seed}', # Directory to save results
        'save_dir': f'save/rbm/mnist/{seed}', # Directory to save results
        'data': 'mnist', # Dataset to use
        'n_steps': 10000, # Number of steps for the sampler
        'n_samples': 500, # Number of samples
        'n_test_samples': 100, # Number of test samples
        'gt_steps': 10000, # Number of steps for ground truth sampling
        'seed': seed, # Random seed
        'n_hidden': 500, # Number of hidden units
        'n_visible': 784, # Number of visible units
        'save_sample_every': 100, # Frequency to save samples
        'print_every': 100, # Print frequency
        'rbm_lr': 0.001, # Learning rate for RBM
        'cd': 10, # Contrastive Divergence steps
        'img_size': 28, # Image size
        'batch_size': 100, # Batch size
        'subsample': 1, # Subsampling factor
        'burn_in': 0.1, # Burn-in period for ESS
        'ess_statistic': 'hamming', # Statistic for ESS
    }
    rbm.rbm_train(args)
    rbm.rbm_sample(args, temps)
```
